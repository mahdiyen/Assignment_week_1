```python
original_glucose_array = np.array(df['glucose'])
interpolated_lin_array = np.array(df['interpolated_lin'].astype('int'))
interpolated_poly_array = np.array(df['interpolated_poly'])
interpolated_pad_array = np.array(df['interpolated_pad'].astype('int'))
```

```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot
from bokeh.plotting import ColumnDataSource
from bokeh.models import DatetimeTickFormatter, NumeralTickFormatter
from datetime import datetime

output_notebook()
#Making a time Array
df['time'] = pd.to_datetime(df['time'])
time_array=np.array(df['time'])

#plotting and comparing original data and linear method od interpolation
p1 = figure(title='comparing to linear method of interpolation ', background_fill_color="white")
p1.line(time_array,interpolated_lin_array, color="blue",legend_label = 'int_pol_lin')
p1.scatter(time_array,original_glucose_array, color="red",legend_label = 'original')
p1.xaxis[0].formatter=DatetimeTickFormatter()
p1.legend.location = "top_left"

#plotting and comparing original data and polynumial method od interpolation
p2 = figure(title='comparing to polynumial method of interpolation', background_fill_color="white")
p2.line(time_array,interpolated_poly_array, color="blue",legend_label = 'int_pol_poly')
p2.scatter(time_array,original_glucose_array, color="red",legend_label = 'original')
p2.xaxis[0].formatter=DatetimeTickFormatter()
p2.legend.location = "top_left"

#plotting and comparing original data and pad method od interpolation
p3 = figure(title='comparing to pad method of interpolation', background_fill_color="white")
p3.line(time_array,interpolated_pad_array, color="blue",legend_label = 'int_pol_pad')
p3.scatter(time_array,original_glucose_array, color="red",legend_label = 'original')
p3.xaxis[0].formatter=DatetimeTickFormatter()
p3.legend.location = "top_left"

p = gridplot([[p1,p2,p3]],width=500, height=300)


show(p)
```

```python
import panel as pn
from bokeh.io import output_notebook
from bokeh.plotting import show, figure
from datetime import datetime
from bokeh.models import DatetimeTickFormatter, NumeralTickFormatter

output_notebook()
pn.extension()
```

```python
def render_plot(x,y,z):
    '''x and y are the x_axis and y axis
    x should be time column
    y should be original data
    z should be one of interpolation methods'''
    y_array = np.array(df[y])
    z_array = np.array(df[z])
    
    df[x] = pd.to_datetime(df[x])
    time_array = np.array(df[x])
    p = figure()
    p.scatter(time_array, y_array,color = 'red', legend_label = 'original')
    p.line(time_array,z_array,color = 'blue', legend_label = 'interpolation')
    p.xaxis[0].formatter = DatetimeTickFormatter()
    return p
```

```python
import pandas as pd
import numpy as np
import json
import yaml

#make a config file
with open('config.yaml','r') as stream:
    config = yaml.safe_load(stream)
file = config['Glucose_file']

#load and read the file
read_df = open(file)
df = json.load(read_df)
df = pd.read_json(df, orient='Records')
df = pd.DataFrame(df)
```

```python
df['interpolated_poly'] = df['glucose'].interpolate(method='polynomial', order=2)
df['interpolated_lin'] = df['glucose'].interpolate(method='linear')
df['interpolated_pad'] = df['glucose'].interpolate(method='pad')
```

```python
columns=df.columns[:]
columns
```

NOTE:As plot shows, rolling mean line method is not suitable. It could be because of having a lot of missing data.

```python
inter_plot = pn.interact(render_plot,x='time',y=columns,z=columns[])
inter_plot
```

```python
dashboard = pn.template.BootstrapTemplate(title='comparing original and interpolated data of glucose',sidebar_width=200)
dashboard.sidebar.append(inter_plot[0])
dashboard.main.append(inter_plot[1])
dashboard.show()
```
